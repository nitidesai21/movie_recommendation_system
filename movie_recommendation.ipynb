## Making User Item Matrix for SVD Model

```python
user_item_matrix=df.pivot_table(
    index="userId",
    columns="movieId",
    values="rating"
).fillna(0)
```

```python
user_item_matrix.head(10)
```

```python
reader=Reader(rating_scale=(1,5))
```

```python
data=Dataset.load_from_df(df[['userId','movieId','rating']],reader)
```

```python
trainset,testset=train_test_split(data,test_size=0.2,random_state=42)
```

```python
svd_model=SVD()
```

```python
svd_model.fit(trainset)
```

```python
predictions = svd_model.test(testset)
```

```python
def recommend_svd(user_id,n=10):
    prediction=[]
    for movie in movies['movieId']:
        predict=svd_model.predict(user_id,movie).est
        prediction.append((movie,predict))
    top_n=sorted(prediction,key=lambda x: x[1],reverse=True)[:n]
    print(f"Top {n} recommendation for User {user_id}:")
    for movie,score in top_n:
        title=movies[movies['movieId']==movie]['title'].values[0]
        print("-",title)
uid=int(input("Enter user id:"))
recommend_svd(uid,n=10)
```

```python
with open('svd_model.pkl', 'wb') as model_file:
    pickle.dump(svd_model, model_file)
print("Model saved as svd_model.pkl")
```

```python
rmse = accuracy.rmse(predictions)
mae = accuracy.mae(predictions)
```

```python
user_item_matrix_array=user_item_matrix.values
```

```python
knn=NearestNeighbors(metric="cosine",algorithm="brute")
knn.fit(user_item_matrix.values)
```

```python
def predictions(knn,user_item_matrix,neighbors=5):
    num_users,num_movies=user_item_matrix.shape
    prediction=np.zeros((num_users,num_movies))
    for i in range(num_users):
        distances,indices=knn.kneighbors(user_item_matrix[i].reshape(1,-1),5)
        neighbors=indices.flatten()[1:]
        prediction[i]=user_item_matrix[neighbors].mean(axis=0)
    return prediction
predicted_rating=predictions(knn,user_item_matrix.values)
```

```python
def recommend_movies(user_id):
    if user_id not in user_item_matrix.index:
        print("User not found")
    user_inx=list(user_item_matrix.index).index(user_id)
    user_ratings=user_item_matrix.iloc[user_inx]
    unrated_movies=np.where(user_ratings==0)[0]
    predicted=predicted_rating[user_inx][unrated_movies]
    top10_movies=unrated_movies[np.argsort(predicted)[::-1][:10]]
    recommend_movie_id=user_item_matrix.columns[top10_movies]
    recommend_titles=movies[movies['movieId'].isin(recommend_movie_id)]['title'].tolist()
    return recommend_titles
```

```python
user_id=1
print(f"Recommended movies for user {user_id}:")
recommendations = recommend_movies(user_id)
for movie in recommendations:
    print(movie)
```

```python
actual=user_item_matrix.values[user_item_matrix.values>0]

predicted_values=predicted_rating[user_item_matrix.values>0]

rmse=np.sqrt(mean_squared_error(actual,predicted_values))
print("RMSE:",rmse)

mae=mean_absolute_error(actual,predicted_values)
print("MAE:",mae)
```
